# Example of using Rise / set and UV coverage

```python
import numpy as np
import matplotlib.pyplot as plt

from scintillation.observatory.observatory import *
from scintillation.observatory.coordinates import *
from astropy.time import Time
from astropy.coordinates import SkyCoord, EarthLocation, AltAz

%matplotlib inline
```

```python
"""
Initialize obs.  

This reads $TEMPO2/observatory/observatories.dat and creates a dictionary
using the long observatory names, and coordinates as an astropy EarthLocation

This version has some custom added observatory locations
"""
obs = Observatories()
obs.allobs
```

# Plot rise / set of a source at one telescope

```python
tel = obs.allobs['PARKES']
t0 = Time('2018-01-01T00:00:00')
source = SkyCoord.from_name('PSRJ0437-4715')

altaz, T = coordinates(t0, source, tel, dT=48., plots=True)

"""
The altitide, azimuth, and time grid are returned, if you want to compute anything
or do custom plotting.
"""
```

# Plot rise / set of a source at many telescopes

```python
tels = obs.lba
t0 = Time('2018-01-01T00:00:00')
source = SkyCoord.from_name('PSRJ0437-4715')

altaz, T = coordinates(t0, source, tels, dT=48., plots=True)
```
